```python
def convertRange(x):
    temp=x.split('-')
    if len(temp)==2:
        return(float(temp[0])+float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
def remove_outliers(data):
    df_output=pd.DataFrame()
    for key,subdf in data.groupby('location'):
        m=np.mean(subdf['price_per_sqft'])
        st=np.std(subdf['price_per_sqft'])
        reduced_df=subdf[(subdf['price_per_sqft']>(m-st)) & (subdf['price_per_sqft']<=(m+st))]
        df_output=pd.concat([df_output,reduced_df],ignore_index=True)
    return df_output
```

```python
def bhk_outliers(data):
    exclude_indices=np.array([])
    for location,location_df in data.groupby('location'):
        bhk_stats={}
        for bhk,bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean':np.mean(bhk_df['price_per_sqft']),
                'std':np.std(bhk_df['price_per_sqft']),
                'count':bhk_df.shape[0]
            }
        # print(location,bhk, bhk_stats[bhk])
        for bhk,bhk_df in location_df.groupby('bhk'):
                stats=bhk_stats.get(bhk-1)
                if stats and stats['count']>5:
                    exclude_indices=np.append(exclude_indices,bhk_df[bhk_df['price_per_sqft']<(stats['mean'])].index.values)
    return data.drop(exclude_indices,axis='index')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
column_trans=make_column_transformer((OneHotEncoder(), ['location']), remainder='passthrough')
```

```python
scaler=StandardScaler()
```

```python
column_trans = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), ['location']),
    (StandardScaler(), ['total_sqft', 'bath', 'bhk']),
    remainder='drop'   # or passthrough if needed
)

pipe = make_pipeline(column_trans, LinearRegression())

pipe.fit(X_train, y_train)
```

```python
y_pred_lr=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

lasso

```python
lasso=Lasso()
```

```python
pipe=make_pipeline(column_trans,lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso=pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_ridge)
```
